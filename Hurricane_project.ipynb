```python
# names of hurricanes
names = ['Cuba I', 'San Felipe II Okeechobee', 'Bahamas', 'Cuba II', 'CubaBrownsville', 'Tampico', 'Labor Day', 'New England', 'Carol', 'Janet', 'Carla', 'Hattie', 'Beulah', 'Camille', 'Edith', 'Anita', 'David', 'Allen', 'Gilbert', 'Hugo', 'Andrew', 'Mitch', 'Isabel', 'Ivan', 'Emily', 'Katrina', 'Rita', 'Wilma', 'Dean', 'Felix', 'Matthew', 'Irma', 'Maria', 'Michael']

# months of hurricanes
months = ['October', 'September', 'September', 'November', 'August', 'September', 'September', 'September', 'September', 'September', 'September', 'October', 'September', 'August', 'September', 'September', 'August', 'August', 'September', 'September', 'August', 'October', 'September', 'September', 'July', 'August', 'September', 'October', 'August', 'September', 'October', 'September', 'September', 'October']

# years of hurricanes
years = [1924, 1928, 1932, 1932, 1933, 1933, 1935, 1938, 1953, 1955, 1961, 1961, 1967, 1969, 1971, 1977, 1979, 1980, 1988, 1989, 1992, 1998, 2003, 2004, 2005, 2005, 2005, 2005, 2007, 2007, 2016, 2017, 2017, 2018]

# maximum sustained winds (mph) of hurricanes
max_sustained_winds = [165, 160, 160, 175, 160, 160, 185, 160, 160, 175, 175, 160, 160, 175, 160, 175, 175, 190, 185, 160, 175, 180, 165, 165, 160, 175, 180, 185, 175, 175, 165, 180, 175, 160]

# areas affected by each hurricane
areas_affected = [['Central America', 'Mexico', 'Cuba', 'Florida', 'The Bahamas'], ['Lesser Antilles', 'The Bahamas', 'United States East Coast', 'Atlantic Canada'], ['The Bahamas', 'Northeastern United States'], ['Lesser Antilles', 'Jamaica', 'Cayman Islands', 'Cuba', 'The Bahamas', 'Bermuda'], ['The Bahamas', 'Cuba', 'Florida', 'Texas', 'Tamaulipas'], ['Jamaica', 'Yucatn Peninsula'], ['The Bahamas', 'Florida', 'Georgia', 'The Carolinas', 'Virginia'], ['Southeastern United States', 'Northeastern United States', 'Southwestern Quebec'], ['Bermuda', 'New England', 'Atlantic Canada'], ['Lesser Antilles', 'Central America'], ['Texas', 'Louisiana', 'Midwestern United States'], ['Central America'], ['The Caribbean', 'Mexico', 'Texas'], ['Cuba', 'United States Gulf Coast'], ['The Caribbean', 'Central America', 'Mexico', 'United States Gulf Coast'], ['Mexico'], ['The Caribbean', 'United States East coast'], ['The Caribbean', 'Yucatn Peninsula', 'Mexico', 'South Texas'], ['Jamaica', 'Venezuela', 'Central America', 'Hispaniola', 'Mexico'], ['The Caribbean', 'United States East Coast'], ['The Bahamas', 'Florida', 'United States Gulf Coast'], ['Central America', 'Yucatn Peninsula', 'South Florida'], ['Greater Antilles', 'Bahamas', 'Eastern United States', 'Ontario'], ['The Caribbean', 'Venezuela', 'United States Gulf Coast'], ['Windward Islands', 'Jamaica', 'Mexico', 'Texas'], ['Bahamas', 'United States Gulf Coast'], ['Cuba', 'United States Gulf Coast'], ['Greater Antilles', 'Central America', 'Florida'], ['The Caribbean', 'Central America'], ['Nicaragua', 'Honduras'], ['Antilles', 'Venezuela', 'Colombia', 'United States East Coast', 'Atlantic Canada'], ['Cape Verde', 'The Caribbean', 'British Virgin Islands', 'U.S. Virgin Islands', 'Cuba', 'Florida'], ['Lesser Antilles', 'Virgin Islands', 'Puerto Rico', 'Dominican Republic', 'Turks and Caicos Islands'], ['Central America', 'United States Gulf Coast (especially Florida Panhandle)']]

# damages (USD($)) of hurricanes
damages = ['Damages not recorded', '100M', 'Damages not recorded', '40M', '27.9M', '5M', 'Damages not recorded', '306M', '2M', '65.8M', '326M', '60.3M', '208M', '1.42B', '25.4M', 'Damages not recorded', '1.54B', '1.24B', '7.1B', '10B', '26.5B', '6.2B', '5.37B', '23.3B', '1.01B', '125B', '12B', '29.4B', '1.76B', '720M', '15.1B', '64.8B', '91.6B', '25.1B']

# deaths for each hurricane
deaths = [90,4000,16,3103,179,184,408,682,5,1023,43,319,688,259,37,11,2068,269,318,107,65,19325,51,124,17,1836,125,87,45,133,603,138,3057,74]

# 1
# Update Recorded Damages
conversion = {"M": 1000000,
              "B": 1000000000}
def damage_conversion (damages):
    converted_damages = []
    for cost in damages:
      if cost == "Damages not recorded":
        converted_damages.append(cost)
      else:
        unit = cost[-1]
        number = float(cost[:-1])
        multiplier = conversion[unit]
        result = number * multiplier
        converted_damages.append(result)
    return converted_damages

print(damage_conversion(damages))


# 2 
# Create a Table
def create_hurricane_table(names, months, years, max_sustained_winds, areas_affected, converted_damages, deaths):
    hurricane_dict = {}
    for i in range(len(names)):
      hurricane_info = {
        "Name": names[i],
        "Month": months[i],
        "Year": years[i],
        "Max sustained Wind": max_sustained_winds[i],
        "Areas Affected": areas_affected[i],
        "Damage": converted_damages[i],
        "Death": deaths[i],
    }
      hurricane_dict[names[i]] = hurricane_info
    return hurricane_dict

print(create_hurricane_table(names, months, years, max_sustained_winds, areas_affected, damages, deaths)["Janet"])


# 3
# Organizing by Year
def create_hurricanes_year(names, months, years, max_sustained_winds, areas_affected, converted_damages, deaths):
    year_dict = {}
    for i in range(len(years)):
        hurricane_info = {
            "Name": names[i],
            "Month": months[i],
            "Year": years[i],
            "Max sustained Wind": max_sustained_winds[i],
            "Areas Affected": areas_affected[i],
            "Damage": converted_damages[i],
            "Death": deaths[i],
        }

        year = years[i]
        if year not in year_dict:
            year_dict[year] = [hurricane_info]
        else:
            year_dict[year].append(hurricane_info)

    return year_dict

print(create_hurricanes_year(names, months, years, max_sustained_winds, areas_affected, damages, deaths)[1932])


# 4
# Counting Damaged Areas
def count_affected_areas(areas_affected):
    area_counts = {}
    for hurricane_areas in areas_affected:
        for area in hurricane_areas:
            if area in area_counts:
                area_counts[area] += 1
            else:
                area_counts[area] = 1
    return area_counts

area_frequencies = count_affected_areas(areas_affected)
print("Times affected: " +str (area_frequencies["Cuba"]))


# 5 
# Calculating Maximum Hurricane Count

def most_affected_area(areas_affected):
    area_counts = count_affected_areas(areas_affected)

    most_affected = ""
    max_count = 0

    for area, count in area_counts.items():
        if count > max_count:
            most_affected = area
            max_count = count

    return (most_affected, max_count)

area, times = most_affected_area(areas_affected)
print(f"The most affected area is {area}, hit {times} times.")


# 6
# Calculating the Deadliest Hurricane

def deadliest_hurricane(names, deaths):
    max_deaths = 0
    deadliest = ""
    for i in range(len(names)):
      if deaths[i] > max_deaths:
        max_deaths = deaths[i]
        deadliest = names[i]
    return(deadliest, max_deaths)

name, death_count = deadliest_hurricane(names, deaths)
print(f"The deadliest hurricane was {name} with {death_count} deaths.")


# 7
# Rating Hurricanes by Mortality
def rate_hurricanes_by_mortality(hurricane_data, mortality_scale):
    rated_hurricanes = {rating: [] for rating in range(6)}  # 0â€“5

    for hurricane in hurricane_data.values():
        deaths = hurricane["Death"]
        rating = 5  # Default to highest

        for scale_rating, upper_bound in mortality_scale.items():
            if deaths <= upper_bound:
                rating = scale_rating
                break

        rated_hurricanes[rating].append(hurricane)

    return rated_hurricanes
    
hurricane_data = create_hurricane_table(names, months, years, max_sustained_winds, areas_affected, damages, deaths)
mortality_scale = {0: 0, 1: 100, 2: 500, 3: 1000, 4: 10000}

rated = rate_hurricanes_by_mortality(hurricane_data, mortality_scale)

for rating, hurricanes in rated.items():
    print(f"Mortality Rating {rating}: {len(hurricanes)} hurricanes")


# 8 Calculating Hurricane Maximum Damage
def most_damaging_hurricane(names, converted_damages):
    max_damage = 0
    worst_hurricane = ""

    for i in range(len(names)):
      if converted_damages[i] != "Damages not recorded" and converted_damages[i] > max_damage:
        max_damage = converted_damages[i]
        worst_hurricane = names[i]
    return worst_hurricane, max_damage

converted_damages = damage_conversion(damages)
name, total_damage = most_damaging_hurricane(names, converted_damages)
print(f"Hurricane {name} cause the maximum damage with a value of {total_damage}$.")


# 9
# Rating Hurricanes by Damage

def rate_hurricanes_by_damage(hurricane_data, damage_scale):
    rated_hurricanes = {rating: [] for rating in range(6)}  # 0â€“5

    for hurricane in hurricane_data.values():
        cost = hurricane["Damage"]

        if isinstance(cost, str):
            continue  # skip this hurricane

        rating = 5  # default to highest
        for scale_rating, upper_bound in damage_scale.items():
            if cost <= upper_bound:
                rating = scale_rating
                break

        rated_hurricanes[rating].append(hurricane)

    return rated_hurricanes
    
hurricane_data = create_hurricane_table(names, months, years, max_sustained_winds, areas_affected, converted_damages, deaths)

damage_scale = {0: 0,
                1: 100000000,
                2: 1000000000,
                3: 10000000000,
                4: 50000000000}

rated = rate_hurricanes_by_damage(hurricane_data, damage_scale)

for rating, hurricanes in rated.items():
    print(f"Damage rating {rating}: {len(hurricanes)} hurricanes")
```
